3. Confusion Matrix merupakan representasi data aktual terhadap data prediksi, dimana hasilnya terdapat 4 matriks, yaitu :
   - True Positive (TP) : Prediksi Benar, Aktual Benar
   - True Negative (TN) : Prediksi Salah, Aktual Benar
   - False Positive (FP) : Prediksi Benar, Aktual Salah
   - False Negative (FN) : Prediksi Salah, Aktual Salah
4. Classification Report adalah akurasi model berdasarkan confusion matriks, yang terdiri dari Precision, Recall, F1-Score.
   - Precision : Nilai TP dibagi dengan jumlah nilai TP dan FP
   - Recall : Nilai TP dibagi dengan jumlah nilai TP dan TN
   - F1-Score : Dua kali Nilai Precision dengan Recall dibagi jumlah Nilai Precision dan Recall.

<h1>Soal 2: Aplikasi Model Evaluasi</h1>

Kali ini kita akan menggunakan data untuk memprediksi kelangsungan hidup pasien yang telah mengalami operasi payudara. Dengan informasi yang dimiliki terkait pasien, kita akan membuat model untuk memprediksi apakah pasien akan bertahan hidup dalam waktu lebih dari 5 tahun atau tidak.

Lebih Lengkapnya kalian bisa membaca informasi tentang dataset di link berikut: https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.names

Buat model Klasifikasi (Model/Algoritma Bebas) untuk memprediksi status pasien dengan ketentuan sebagai berikut:

1. Bagi kedua data ini menjadi data training dan data test dengan test_size=0.25.
3. Pelajar tentang metrics roc_auc_score kemudian buatlah model dan evaluasi dengan menggunakan teknik cross-validation dengan scoring 'roc_auc'. Baca https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html untuk menggunakan metric roc_auc saat cross-validation.
3. Berapa score rata2 dari model dengan teknik cross-validation tersebut?
4. Prediksi data test dengan model yang telah kalian buat!
5. Bagaimana hasil confusion matrix dari hasil prediksi tersebut?
6. Bagaimana classification report dari hasil prediksi tersebut?
5. Seberapa baik model anda dalam memprediksi seorang pasien mempunyai status positive?
6. Seberapa baik model anda dalam memprediksi seorang pasien mempunyai status negatif?

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.csv'
list_cols = ['Age', "Patient's Years", "N_positive_ax", "survival_status"]
df = pd.read_csv(url, names=list_cols)
```

```python
df.head()
```

```python
df['survival_status'].value_counts()
```

```python
# Code here
# Pembagian Data Train dan Test
from sklearn.model_selection import train_test_split

x = df.drop('survival_status', axis=1)
y = df['survival_status']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=21)

# Model Prediksi
# Algoritma KNN
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 6, weights = 'distance')
model.fit(x_train, y_train)
y_prediksi = model.predict(x_test)

# K-Fold Cross Validation
from sklearn.model_selection import cross_val_score
print('Rata-rata K-Fold Cross Validation :', cross_val_score(model, x, y, scoring='roc_auc', cv=5).mean())

# Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, y_prediksi)
print('Confusion Matrix : \n', confusion)

# Classification Report 
from sklearn.metrics import classification_report
print('Hasil Classification Report :\n', classification_report(y_test, y_prediksi, target_names=['Survived', 'Died']))

# Prediksi Status Pasien
# Nilai Matrix
TP = confusion[0][0]
FP = confusion[0][1]
TN = confusion[1][0]
FN = confusion[1][1]
# Status Positive
precision_P = TP / (TP + FP)
recall_P = TP / (TP + TN)
f1_score_P = 2 * precision_P * recall_P / (precision_P + recall_P)
print('Akurasi Prediksi Pasien Berstatus Positive :', round(f1_score_P, 2))
# Status Negative
precision_N = FN / (TN + FN)
recall_N = FN / (FN + FP)
f1_score_N = 2 * precision_N * recall_N / (precision_N + recall_N)
print('Akurasi Prediksi Pasien Berstatus Negative :', round(f1_score_N, 2))
```

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.csv'
list_cols = ['Age', "Patient's Years", "N_positive_ax", "survival_status"]
df = pd.read_csv(url, names=list_cols)
```

```python
df.head()
```

```python
# Code here
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, cross_validate, RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score

# Pembagian Data Train dan Test
x = df.drop('survival_status', axis=1)
y = df['survival_status']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=21)
```

```python
# Model Prediksi Algoritma KNN with GridSearchCV
model_knn = KNeighborsClassifier()
params = {'n_neighbors':np.arange(5, 80),
          'weights':['uniform', 'distance']}
gscv = GridSearchCV(model_knn, param_grid=params, scoring='roc_auc', cv=5)
gscv.fit(x_train, y_train)
print('Hyperparameter Terbaik :', gscv.best_params_)
print('Score GridSearchCV :', gscv.best_score_)

# Scoring Validation
validate = cross_validate(gscv, x, y, cv=5, return_train_score=True)
print('Train Score Validasi :', validate['train_score'].mean())
print('Test Score Validasi :', validate['test_score'].mean())

# Probabilitas
probabilitas = gscv.predict_proba(x_test)
print('Hasil Probabilitas Pasien Positif Meninggal', probabilitas[1][0])

# Scoring ROC AUC
print('Hasil Scoring Model', roc_auc_score(y_test, gscv.predict(x_test)))
```

Model KNN (K-Nearest Neighbors) yang dihasilkan mungkin cukup baik walaupun akurasi tidak maksimal, terdapat perbedaan scoring validasi sekitar 5% antara data train dan data test.

```python
# Model Prediksi Decision Tree with RandomizedSearchCV
model_tree = DecisionTreeClassifier()
params = {'max_depth':np.arange(5, 50),
          'min_samples_leaf':np.arange(0.1, 0.5)}
randomcv = RandomizedSearchCV(model_tree, param_distributions = params, 
                              scoring='roc_auc', n_iter=20, cv=5)
randomcv.fit(x_train, y_train)
print('Hyperparameter Terbaik :', randomcv.best_params_)
print('Score RandomizedSearchCV :', randomcv.best_score_)

# Scoring Validation
validate = cross_validate(randomcv, x, y, cv=5, return_train_score=True)
print('Train Score Validasi :', validate['train_score'].mean())
print('Test Score Validasi :', validate['test_score'].mean())

# Scoring ROC AUC
print('Hasil Scoring Model', roc_auc_score(y_test, randomcv.predict(x_test)))
```
