```
def ambil_duit(duit, dompetnya):
    dompetku=0
    for uang in dompetnya :
      if dompetku == duit:
        return
      dompetku += 1
      yield uang

def satuaja(dompetnya):
  didompetku = set()
  for uang in dompetnya:
    if uang in didompetku:
      continue
    yield uang
    didompetku.add(uang)

def ambilkan(lembar):
  uang = [5,10,5,20,10,50,10,1,5,2,100]
  for duit in ambil_duit(lembar, satuaja(uang)):
    print(duit, end=' ')

ambilkan(5)
```

```
makanan=[]
def yoyoyo():
  makanan.append('ayam')
  return makanan

def ayayay():
  makanan.pop(200)
  return makanan

def yeyeye():
  makanan.extend(['dimsum'])
  return makanan

def ayeayeaye():
  makanan.insert(10,'batagor')
  return makanan

for _ in range(1000):
  yoyoyo()
for _ in range(3):
  ayayay()
for _ in range(20):
  yeyeye()
for _ in range(2):
  ayeayeaye()

from collections import Counter
print(Counter(makanan))
```
